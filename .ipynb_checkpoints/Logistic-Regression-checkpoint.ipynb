# Mapping Loan Grades

This section transforms the `grade` feature into ordinal values, where 'A' is mapped to 1, 'B' to 2, and so forth, up to 'G', which is mapped to 7. Unique values and their counts are also examined to understand the distribution of loan grades in the dataset.




```python
ordinal_mapping = {'A':1, 'B':2, 'C':3, 'D':4, 'E':5 ,'F':6 ,'G':7}
data['grade'] = data['grade'].map(ordinal_mapping)
data['grade'].unique(),data['grade'].value_counts()
```

# Encoding Sub-Grades

This section focuses on encoding the `sub_grade` feature for analysis. The process involves the following steps:

1. **Extract Letter and Numeric Parts**: The `sub_grade` is split into its letter and numeric components.
2. **Map Letter Grades**: Letter grades are mapped to numeric values (A=1, B=2, ..., G=7).
3. **Combine Encodings**: The letter and numeric values are combined to create a single `encoded_score`.
4. **Reverse the Score**: Lower numbers represent higher ranks, so the score is adjusted accordingly.
5. **Update Sub-Grade**: The original `sub_grade` is replaced with the new encoded score.

Finally, the unique values of the updated `sub_grade` are examined to assess the new encoding.


```python
data['sub_grade'].unique()
```

```python
letter_grade_mapping ={'A':1, 'B':2, 'C':3, 'D':4, 'E':5 ,'F':6 ,'G':7}
# Extracting letter grade and numeric sub-grade
data['letter_grade'] = data['sub_grade'].str[0]
data['numeric_sub_grade'] = data['sub_grade'].str[1:].astype(int)

# Maping letter grades to numbers
data['letter_encoded'] = data['letter_grade'].map(letter_grade_mapping)

# Combining letter and numeric encoding
data['encoded_score'] = data['letter_encoded'] * 10 + data['numeric_sub_grade']

# Want lower numbers to be earliest_cr_line :The month the borrower's earliest reported credit line was openedhigher rank so reverse the score
data['encoded_score'] = data['encoded_score'].max() - data['encoded_score'] + 1

data['sub_grade'] = data['encoded_score']

data = data.drop(columns=['letter_grade', 'numeric_sub_grade', 'letter_encoded', 'encoded_score'])

data['sub_grade'].unique()
```

# Updating Employment Title Data with Loan Repayment Proportions

In this section, the employment title data is enhanced to reflect the proportions of 'Fully Paid' loans. The following steps were taken:

1. **Counted Occurrences**: The total number of occurrences for each employment title was calculated.
2. **Filtered Titles**: Titles with more than 100 occurrences were retained for analysis.
3. **Calculated Proportions**: The proportion of 'Fully Paid' loans for each employment title was computed.
4. **Mapped Proportions**: The employment titles in the dataset were replaced with their corresponding repayment proportions, using a defined function to ensure that titles outside the filtered list received a value of 0.

The updated dataset now includes proportions of loan repayment success associated with each employment title.


```python
total_by_emp_title = data['emp_title'].value_counts()

filtered_titles = total_by_emp_title[total_by_emp_title > 100].index

fully_paid_by_emp_title = data[data['loan_status'] == 'Fully Paid']['emp_title'].value_counts()
proportion_fully_paid = (fully_paid_by_emp_title / total_by_emp_title).fillna(0)

proportion_fully_paid_filtered = proportion_fully_paid[proportion_fully_paid.index.isin(filtered_titles)]

def replace_with_proportion(emp_title):
    if emp_title in proportion_fully_paid_filtered.index:
        return proportion_fully_paid_filtered[emp_title]
    else:
        return 0

data['emp_title'] = data['emp_title'].apply(replace_with_proportion)

data.head(5)
```

# Outlier Detection and Removal

1. **Z-Score Method**: Identifying outliers using Z-scores with a threshold of 3.5.
2. **IQR Method**: Using the Interquartile Range to detect outliers beyond 1.5 times the IQR.
3. **Common Outliers**: Finding and removing common outliers identified by both methods.
4. **Data Size**:
   - Original: 396,030  
   - Common Outliers: 22,306  
   - Cleaned: 373,724  


```python
# Step 1: Z-Score Method
# Calculating Z-Scores for each column
z_scores = stats.zscore(df)
z_scores_df = pd.DataFrame(z_scores, columns=df.columns)
z_threshold = 3.5
# Identifying outliers based on Z-scores
z_outliers = (z_scores_df.abs() > z_threshold)
z_outlier_indices = z_outliers[z_outliers.any(axis=1)].index

# Step 2: IQR Method
# Calculating Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
iqr_threshold = 1.5
iqr_outliers = ((df < (Q1 - iqr_threshold * IQR)) | (df > (Q3 + iqr_threshold * IQR)))
iqr_outlier_indices = iqr_outliers[iqr_outliers.any(axis=1)].index
# Finding common outliers
common_outlier_indices = z_outlier_indices.intersection(iqr_outlier_indices)
# Printing the size of common outliers and original data
print("Common outliers size identified by both methods and original data size :")
print(common_outlier_indices.shape, df.shape)
# Dropping the common outliers from the DataFrame
df_cleaned = df.drop(index=common_outlier_indices, errors='ignore')

df_cleaned.reset_index(drop=True, inplace=True)

print("Data size after dropping outliers: ")
print(df_cleaned.shape)
```

# Chi-Square Test on grade and loan status
### Hypotheses

- **Null Hypothesis (H0)**: There is no significant relationship between loan grade and loan payment status (fully paid vs. charged off).

- **Alternative Hypothesis (H1)**: There is a significant relationship between loan grade and loan payment status (fully paid vs. charged off).

### Result

- **p-value**: (Insert p-value here)
- **Conclusion**: Since the p-value is less than the significance level (Î± = 0.05), we **reject the null hypothesis**. This indicates that there is a significant relationship between loan grade and loan payment status.


```python
contingency_table = pd.crosstab(df['grade'], df['loan_status'])

# Performing Chi-Square test
chi2_stat, p_val, dof, ex = chi2_contingency(contingency_table)

#p-value
alpha = 0.05
if p_val < alpha:
    result = "Reject the null hypothesis. There is a significant relationship between grade and loan payment status."
else:
    result = "Fail to reject the null hypothesis. There is no significant relationship between grade and loan payment status."

# Calculating number of fully paid loans and total loans for grade 'A'
grade_a_total = df[df['grade'] == 1].shape[0]
grade_a_fully_paid = df[(df['grade'] == 1) & (df['loan_status'] == 0)].shape[0]

# Calculatingproportion and percentage
grade_a_fully_paid_percentage = (grade_a_fully_paid / grade_a_total) * 100 if grade_a_total > 0 else 0

print(f"Percentage of fully paid loans among grade 'A' borrowers: {grade_a_fully_paid_percentage:.2f}%")

print(result)
```

# Analysis of Fully Paid Loans by Grade

 Calculating the percentage of fully paid loans for each loan grade (A to G). It first defines a mapping of grades to numerical values for ease of computation and then reverses this mapping for interpretation. The code iterates through each unique grade, calculates the total number of loans and the number of fully paid loans for that grade, and computes the percentage of fully paid loans. Finally, it prints the percentages for each grade.


```python
ordinal_mapping = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5 ,'F':6,'G':7}

# reverse mapping from numbers back to grades
reverse_mapping = {v: k for k, v in ordinal_mapping.items()}

# Percentage of fully paid loans for each grade
grade_fully_paid_percentages = {}


unique_grades = df['grade'].unique()
for grade in unique_grades:
    total_loans = df[df['grade'] == grade].shape[0]
    fully_paid_loans = df[(df['grade'] == grade) & (df['loan_status'] == 0)].shape[0]
    fully_paid_percentage = (fully_paid_loans / total_loans) * 100 if total_loans > 0 else 0
    #reverse mapping to get the letter grade
    grade_letter = reverse_mapping.get(grade, 'Unknown')
    grade_fully_paid_percentages[grade_letter] = fully_paid_percentage

for grade_letter, percentage in grade_fully_paid_percentages.items():
    print(f"Percentage of fully paid loans for grade '{grade_letter}': {percentage:.2f}%")
```

# ANOVA Analysis of Annual Income by Loan Status

This code performs a one-way ANOVA test to analyze whether there are significant differences in annual income among different loan statuses. The test calculates the F-statistic and the associated p-value. Based on the p-value, it determines whether to reject or fail to reject the null hypothesis.

### Results:
- **F-statistic:** 2258.4821
- **P-value:** 0.0000

### Conclusion:
- Since the p-value is less than 0.05, we reject the null hypothesis, indicating there is a significant difference in annual income among the different loan statuses.


```python
# Performing ANOVA
anova_results = stats.f_oneway(*[df[df['loan_status'] == status]['annual_inc'] for status in df['loan_status'].unique()])

print(f"F-statistic: {anova_results.statistic:.4f}")
print(f"P-value: {anova_results.pvalue:.4f}")

alpha = 0.05
if anova_results.pvalue < alpha:
    print("Reject the null hypothesis: There is a significant difference in annual income among the different loan statuses.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in annual income among the different loan statuses.")
```

# Evaluating Logistic Regression with Varying Regularization

Training and validation scores are computed for a range of regularization parameters (Î») in a logistic regression model. The results are plotted to visualize the impact of regularization on model accuracy.

### Observations:
- The plot illustrates how accuracy changes with different values of Î» for both training and validation sets.
- A balance between train and validation scores helps identify the optimal regularization parameter.


```python
train_scores = []
val_scores = []
scaler = StandardScaler()

for la in np.arange(0.01, 5000.0, 100): # range of values of Lambda

  scaled_lr = make_pipeline(scaler, LogisticRegression(C=1/la))
  scaled_lr.fit(X_train, y_train)
  train_score = accuracy(y_train, scaled_lr.predict(X_train))
  val_score = accuracy(y_test, scaled_lr.predict(X_test))
  train_scores.append(train_score)
  val_scores.append(val_score)
```

```python
plt.figure(figsize=(10,5))
plt.plot(list(np.arange(0.01, 5000.0, 100)), train_scores, label="train")
plt.plot(list(np.arange(0.01, 5000.0, 100)), val_scores, label="val")
plt.legend(loc='lower right')

plt.xlabel("Regularization Parameter(Î»)")
plt.ylabel("Accuracy")
plt.grid()
plt.show()
```

# Evaluating Logistic Regression with Polynomial Features

This analysis investigates how the degree of polynomial features impacts the performance of a logistic regression model on a given dataset.

```python
from sklearn.pipeline import make_pipeline
degree = 3

for d in range(degree):

  model = make_pipeline(PolynomialFeatures(degree=d), StandardScaler(),
                        LogisticRegression(C=1/10,max_iter=300,random_state=0))

  model.fit(X_train, y_train)

  print('*'*50)
  print("Degree:", d)
  print("Training Accuracy:", accuracy(y_train, model.predict(X_train)))
  print("Test Accuracy:", accuracy(y_test, model.predict(X_test)))
  print("Training Recall:", recall_score(y_train, model.predict(X_train)))
  print("Test Recall:", recall_score(y_test, model.predict(X_test)))
  print("Training Precision:", precision_score(y_train, model.predict(X_train)))
  print("Test Precision:", precision_score(y_test, model.predict(X_test)))
```

```python
model = make_pipeline(PolynomialFeatures(degree=2), StandardScaler(),
                      LogisticRegression(C=1/10, max_iter=300, random_state=0))

# Train the model
model.fit(X_train, y_train)

# Get the predicted probabilities
probability = model.predict_proba(X_test)
probabilities = probability[:, 1]

# Compute FPR, TPR, and thresholds
fpr, tpr, thresholds = roc_curve(y_test, probabilities)

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label='ROC curve')
plt.plot(fpr, fpr, '--', color='red', label='Random model')
plt.title('ROC curve')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')

# Annotate some threshold values on the plot
for i, thr in enumerate(thresholds):
    # Annotate at every nth point to avoid clutter, adjust as necessary
    if i % 20 == 0:
        plt.annotate(f'{thr:.2f}', (fpr[i], tpr[i]), fontsize=8, color='blue')

# Show the legend and plot
plt.legend()
plt.show()
```

```python
precision, recall, thresholds = precision_recall_curve(y_test, probabilities)

# Plot the Precision-Recall curve
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, label='Precision-Recall curve')
plt.title('Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')

# Annotate some threshold values on the curve
thresholds = np.append(thresholds, 1)  # Append 1 for the last threshold point
for i, thr in enumerate(thresholds):
    # Annotate at every nth point to avoid clutter, adjust as necessary
    if i % 20 == 0:
        plt.annotate(f'{thr:.2f}', (recall[i], precision[i]), fontsize=8, color='blue')

# Show the plot
plt.legend()
plt.show()
```
